```python
df2 = df2.dropna()
df2
```

```python
df2['cs_status-wc-label'].loc[df2['cs_status-wc-label'] == 'Conversão'] = 1
df2['cs_status-wc-label'].loc[df2['cs_status-wc-label'] == 'Recusa'] = 0
df2
```

```python
#df2.drop(columns = ['birth_date'])
df2['income'] = df2['income'].str.replace(',','.')
df2['limit'] = df2['income'].str.replace(',','.')
df2['income'] = df2['income'].astype(float)
df2['limit'] = df2['limit'].astype(float)

df2.drop(columns = ['birth_date'], inplace = True)
y = df2['cs_status-wc-label'].loc[(df2['cs_status-wc-label'] == 1) | (df2['cs_status-wc-label'] == 0)]
df2.drop(columns  = ['cs_status-wc-label'] , inplace = True)
df_final = pd.get_dummies(data = df2,columns = ['sex','marital_status','state','offer','step_abandonment'],drop_first=True)
```

```python
from imblearn.over_sampling import RandomOverSampler
```

```python
y = y.astype(int)
y.value_counts()
```

```python
from imblearn.over_sampling import SMOTE
```

```python
oversample = SMOTE(random_state= 101)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
X_train,X_teste,y_train,y_test = train_test_split(df_final,y, test_size = 0.33, random_state = 42)
```

```python
x_train,y_train1 = oversample.fit_resample(X_train,y_train)
```

```python
log = LogisticRegression()
```

```python
log.fit(x_train,y_train1)
z = log.predict(X_teste)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
print(accuracy_score(y_test,z))
```

```python
print(confusion_matrix(y_test,z))
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(z,y_test))
```
